## 1d. Clean Date Col

The start_date and end_date do not have standardise date format and is of mixed data of format like %Y-%m-%d, %Y-%m and %y/%m/%d, so would need to standardise them so that it would be easier for me to convert it to a datetime column

```python
import pandas as pd

# Attempt to convert dates using common formats including both '%d/%m/%Y' and '%d/%m/%y'
def try_multiple_formats(date_str):
    formats = [
    "%d/%m/%Y", "%d-%m-%Y", "%d.%m.%Y",
    "%m/%d/%Y", "%m-%d-%Y", "%m.%d.%Y",
    "%Y/%m/%d", "%Y-%m-%d", "%Y.%m.%d",
    "%Y/%m", "%Y-%m", "%Y.%m",
    "%d/%m/%y", "%d-%m-%y", "%d.%m.%y",
    "%m/%d/%y", "%m-%d-%y", "%m.%d.%y",
    "%y/%m/%d", "%y-%m-%d", "%y.%m.%d",
    "%y/%m", "%y-%m", "%y.%m",
    "%b %d, %Y", "%B %d, %Y", "%d %b, %Y", "%d %B, %Y",
    "%d %b %Y", "%d %B %Y",  # Formats with month names
    "%Y",  # Just the year
    "%b %Y", "%B %Y",  # Month name and year
]  
    # List of formats to try
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt, errors='raise')
        except ValueError:
            continue
    return None # Return NaT if all formats fail

# Apply the conversion function
anime['start_date'] = anime['start_date'].apply(try_multiple_formats)
anime['end_date'] = anime['end_date'].apply(try_multiple_formats)

# Check entries that failed to convert
print("Failed conversions:", anime['start_date'].isna().sum())
print("Failed conversions:", anime['end_date'].isna().sum())
```

```python
# Function to determine the season from month
def get_season(month):
    if 1 <= month <= 3:
        return 'winter'
    elif 4 <= month <= 6:
        return 'spring'
    elif 7 <= month <= 9:
        return 'summer'
    elif 10 <= month <= 12:
        return 'fall'

# Apply this function to fill null values in the 'season' column based on 'start_date' month
anime.loc[anime['season'].isnull(), 'season'] = anime['start_date'].dt.month.apply(get_season)
```

```python
def map_season(month):
    if 1 <= month <= 3:
        return 'winter'
    elif 4 <= month <= 6:
        return 'spring'
    elif 7 <= month <= 9:
        return 'summer'
    elif 10 <= month <= 12:
        return 'fall'
    else:
        return None  # Just in case there are any out-of-range months
        
anime['season'] = anime['start_date'].dt.month.apply(map_season)
```

## 2a. Split Statistics Column Up

```python
def split_statistics_col(clean_data):
    # Convert the 'statistics' column from JSON string to dictionary
    clean_data['statistics'] = clean_data['statistics'].apply(lambda x: json.loads(x.replace("'", "\"")))

    # Extract clean_data from the dictionary and assign it to new columns
    clean_data['statistics_watching'] = clean_data['statistics'].apply(lambda x: x['status']['watching'])
    clean_data['statistics_completed'] = clean_data['statistics'].apply(lambda x: x['status']['completed'])
    clean_data['statistics_on_hold'] = clean_data['statistics'].apply(lambda x: x['status']['on_hold'])
    clean_data['statistics_dropped'] = clean_data['statistics'].apply(lambda x: x['status']['dropped'])
    clean_data['statistics_plan_to_watch'] = clean_data['statistics'].apply(lambda x: x['status']['plan_to_watch'])
    clean_data['statistics_num_list_users'] = clean_data['statistics'].apply(lambda x: x['num_list_users'])

    # Drop the original 'statistics' column
    clean_data.drop(columns=['statistics'], inplace=True)

    return clean_data
```

## 2b. Featuring Engineering with Statistic Column

```python
anime['statistics_watching'] = anime['statistics_watching'].astype(int)
anime['statistics_completed'] = anime['statistics_completed'].astype(int)
anime['statistics_dropped'] = anime['statistics_dropped'].astype(int)
anime['statistics_on_hold'] = anime['statistics_on_hold'].astype(int)
```

```python
def calculate_rates(df):

    # Ensure columns exist to avoid KeyError
    required_columns = ['statistics_watching', 'statistics_completed', 'statistics_on_hold', 'statistics_dropped']
    if not all(col in df.columns for col in required_columns):
        raise ValueError("Missing one or more required columns in the DataFrame")

    # Calculate rates
    df['completion_rate'] = df.apply(lambda row: row['statistics_completed'] / 
        (row['statistics_watching'] + row['statistics_completed'] + row['statistics_dropped']) if row['statistics_watching'] + row['statistics_completed'] + row['statistics_dropped'] > 0 else 0, axis=1)

    df['drop_rate'] = df.apply(lambda row: row['statistics_dropped'] / 
        (row['statistics_watching'] + row['statistics_completed'] + row['statistics_dropped']) if row['statistics_watching'] + row['statistics_completed'] + row['statistics_dropped'] > 0 else 0, axis=1)

    df['retention_rate'] = df.apply(lambda row: (row['statistics_completed'] + row['statistics_watching'] + row['statistics_on_hold']) / 
        (row['statistics_watching'] + row['statistics_completed'] + row['statistics_on_hold'] + row['statistics_dropped']) if row['statistics_watching'] + row['statistics_completed'] + row['statistics_on_hold'] + row['statistics_dropped'] > 0 else 0, axis=1)

    return df

calculate_rates(anime)
```

## 3.1 Time Series Dataset 1980s Onwards

We can see that from year 1917-05 to 2024-03, but the infront is not continuous as some of the months are missing in between. Therefore we decide to only visualise the 2000s anime. 

Using data from the 1980s onward to analyze anime genre trends can be advantageous for several reasons. First, the global popularity of anime has seen significant growth since the 1980s due to advancements in technology, which has improved access through online streaming platforms, thus making the data more representative of current viewership patterns. Moreover, cultural shifts and changes in audience preferences are better reflected in more recent data, providing a relevant snapshot for current market trends.

Secondly, the anime industry has evolved with the introduction of new genres and subgenres, and earlier data might not adequately capture these nuances. Trends observed in the last two decades are more likely to continue into the future, making them more actionable.

Regarding the use of a non-continuous index for time series, it can lead to misleading results and analytical complications. Time series analysis relies on the assumption that data points occur at regular intervals to correctly model and forecast future trends. Missing time points can disrupt patterns and seasonality, leading to inaccurate trend analysis and forecasting. Therefore, ensuring a continuous time index without gaps is crucial for maintaining the integrity of time series models and analyses.

**Conclusion:** Therefore, I will only be using data from 1980 onwards, but since there still a few missing months in between I will be filling it myself

# 4. Create Source Time Series DataFrame

```python
#convert source trend to time series df
def create_source_time_series(df):
   
    # Convert 'start_date' to datetime and extract year-month
    df['start_date'] = pd.to_datetime(df['start_date'])
    df['year_month'] = df['start_date'].dt.to_period('M')

    # Explode the source column into separate rows
    exploded_df = df.explode('source')

    # Group by 'year_month' and 'source', count occurrences, and pivot sources to columns
    source_ts = exploded_df.groupby(['year_month', 'source']).size().unstack(fill_value=0)

    # Reindex to fill in missing months
    min_year_month = source_ts.index.min()
    max_year_month = source_ts.index.max()
    full_range = pd.period_range(start=min_year_month, end=max_year_month, freq='M')
    source_ts = source_ts.reindex(full_range, fill_value=0)

    # Filter out data before 1980
    source_ts = source_ts[(source_ts.index >= '1980-01')]

    # Reset the index
    source_ts.reset_index(inplace=True)
    source_ts.columns.name = None
    source_ts.rename(columns={'index': 'year_month'}, inplace=True)

    # Return the filtered and complete time series DataFrame
    return source_ts

# Use the function to create the source time series DataFrame
source_ts_df = create_source_time_series(anime_clean)

# Print the resulting DataFrame
source_ts_df
```
